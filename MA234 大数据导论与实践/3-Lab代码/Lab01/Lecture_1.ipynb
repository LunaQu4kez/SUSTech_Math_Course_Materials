### Dicts

```python
# 
dict1 = {'Marshall Medical Center South':0.98,'Wedowee Hospital':0.84}
dict2 = dict(Marshall_Medical_Center_South=0.98,Wedowee_Hospital=0.84)
dict3 = dict([('Marshall Medical Center South',0.98),('Wedowee Hospital',0.84)])

dict4 = dict([('Wedowee Hospital',0.84),('Marshall Medical Center South',0.98)]) # 
```

```python
print(dict1)
print(dict2)
print(dict3)
print(dict4)
```

```python
# 
from collections import OrderedDict
ordered_dict1 = OrderedDict([('Wedowee Hospital',0.84),('Marshall Medical Center South',0.98)])
ordered_dict2 = OrderedDict([('Marshall Medical Center South',0.98),('Wedowee Hospital',0.84)])
print(ordered_dict1)
print(ordered_dict2)
```

```python
# 
key_list = ['Marshall Medical Center South','Wedowee Hospital']
value_list = [0.98,0.84]
dict5 = {}
for key, value in zip(key_list,value_list):
    dict5[key] = value
print(dict5)
```

```python
# 

print(dict5['Marshall Medical Center South']) # 
print(dict5.get('Marshall Medical Center South')) # get
print(dict5.get('Marshall')) # none
print(dict5.get('Marshall',u'')) # 
print('Marshall Medical Center South' in dict5) # in 
print(dict5.__contains__('Marshall')) # has_key 
```

```python
print(type(dict5.keys())) # <class 'dict_keys'>for
print(dict5.keys()) # 
print(dict5.values()) # 
print(dict5.items())# 
```

```python
for key,value in dict5.items(): # items
    print(key,":",value)
```

```python
dict5['City Hospital']= 1 #  
print(dict5)
del dict5['Wedowee Hospital'] # del
print(dict5) 
dict5.pop('City Hospital') # pop
print(dict5)
dict5.clear() # 
print(dict5)
```
